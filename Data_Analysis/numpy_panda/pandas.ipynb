```python
df = pd.read_csv('data.csv')
df.head(3)
```

```python
df['new_val'] = df['Value'].fillna(df['Value'].mean())
df
```

```python
grp_mean = df.groupby('Product')['Value'].mean()
print(grp_mean)
grp_sum = df.groupby(['Product','Region'])['Value'].sum()
print(grp_sum)
```

```python
grp_agg = df.groupby('Region')['Value'].agg(['sum','count','mean'])
grp_agg
```

```python
# merging and joining
df1 = pd.DataFrame({'keys': ['a','b','c'], 'value1':[1,2,3]})
df1
df2 = pd.DataFrame({'keys':['a','b','d'], 'value2':[2,3,4]})
df2
```

```python
pd.merge(df1,df2,on="keys",how="inner")
```

```python
pd.merge(df1,df2,on="keys",how="outer")
```

```python
pd.merge(df1,df2,on="keys",how="left")
```
