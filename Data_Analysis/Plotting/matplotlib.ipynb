```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,1,29,39,32]
plt.plot(x,y)
```

```python
plt.Figure(figsize=(9,5))

plt.subplot(1,2,1)
plt.plot(x,y,color='red',linewidth=2)

plt.subplot(1,2,2)
plt.plot(y,x,color='green')
```

```python
plt.bar(x,y)
```

```python
plt.hist(x,bins=4,color='red',edgecolor='black')
```

```python
val = [10,20,30,50]
labels = ['a','b','c','d']
explode = (0.3,0,0,0)
plt.pie(val,labels = labels,explode=explode,autopct="%1.1f%%")
```
