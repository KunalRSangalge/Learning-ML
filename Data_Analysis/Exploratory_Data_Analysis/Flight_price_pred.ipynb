```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_excel('flight_price.xlsx')
df.head()
```

```python
# we can see the data is very raw, we will clean the data row by row now
df.info()
```

```python
df.describe()
```

```python
#date 
df['Date'] = df['Date_of_Journey'].str.split('/').str[0] #date
df['Month'] = df['Date_of_Journey'].str.split('/').str[1] #month
df['Year'] = df['Date_of_Journey'].str.split('/').str[2] #year
df.head()
```

```python
df.info()
```

```python
#date,month,year are still object type convert in numerical
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
df.info()
```

```python
df.drop(columns='Date_of_Journey',inplace=True)
```

```python
df.info() #successfully done date
```

```python
#arrival_time , we can see that 22 Mar are not required so first we will remove them and then make the Arrival_hour, Arrival_min
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
df.head()
```

```python
#arrival_time
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
df.head()
```

```python
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
df.info()
```

```python
df.drop(columns='Arrival_Time',inplace=True)
```

```python
df.info()
```

```python
#dept_time
df['Departure_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Departure_min'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df['Departure_hour']=df['Departure_hour'].astype(int)
df['Departure_min']=df['Departure_min'].astype(int)
```

```python
df.drop(columns='Dep_Time',inplace=True)
df.head()
```

```python
#duration
df['Duration_hour'] = df['Duration'].str.extract(r'(\d+)h').fillna(0).astype(int)
df['Duration_min'] = df['Duration'].str.extract(r'(\d+)m').fillna(0).astype(int)
df.info()
```

```python
df.drop(columns='Duration',inplace=True)
```

```python
##Taking categorical Features - stops
df['Total_Stops'].unique() 
```

```python
#we will map stops with numbers cause we know as stops increase, price increases
mapping = {'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3,'4 stops':4}
df['Total_Stops_numeric'] = df['Total_Stops'].map(mapping).fillna(1) #did something wrong with excel but this is correct        
```

```python
#drop route becuase we have sorce and destination
df.drop(columns='Route',inplace=True)
df.info()
```
